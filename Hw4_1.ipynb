```python
from scipy import integrate
import numpy as np
```

```python
speed = [25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65]
time = np.linspace(0,11,12)
print(time)
```

```python
import matplotlib.pyplot as plt

plt.plot(
    time,
    speed,
    color="#061358",
    label="speed",
    linewidth=1,
    marker="*",
)
plt.xlim(0, 11)
plt.ylim(0, 130)
plt.xlabel("Час", fontsize="small", color="midnightblue")
plt.ylabel("Скорость", fontsize="small", color="midnightblue")
plt.title("Значения скорости в определенный момент наблюдения", fontsize=15)
plt.legend()
plt.grid()
plt.show()
```

```python
from scipy.interpolate import interp1d

f = interp1d(time, speed, kind='cubic')

plt.plot(time, speed, 'o')
more_time = np.linspace(0, 11, 1000)
plt.plot(more_time, f(more_time))
plt.ylim(0,130)
```

```python
from scipy import integrate

result = integrate.quad(lambda x: f(x), 0, 11)
print(result)
```

```python
f_1 = interp1d(time, speed, kind='quadratic')
plt.plot(time, speed, 'o')
more_x = np.linspace(0, 11, 1000)
plt.plot(more_x, f(more_x))
```

```python
result = integrate.quad(lambda x: f_1(x), 0, 11)
print(result)
```
