```python
import matplotlib.pyplot as plt
import random

x = []
y = []

Ax = 30.0
Ay = 0.0
Bx = -30.0
By = 0.0
Cx = 0.0
Cy = 100
x.append(Ax)
y.append(Ay)
x.append(Bx)
y.append(By)
x.append(Cx)
y.append(Cy)
plt.figure(figsize=(16,10), dpi= 80)
plt.plot(x,y, 'bo')
plt.show()
```

```python
pX = 0.0
pY = 0.0
for i in range(0, 5000):
    x.append(pX)
    y.append(pY)
    C = random.randint(0,3)
    if C == 0:
       pX = pX + (Ax - pX) / 2
       pY = pY + (Ay - pY) / 2
    if C == 1:
       pX = pX + (Bx - pX) / 2
       pY = pY + (By - pY) / 2
    if C == 2:
       pX = pX + (Cx - pX) / 2
       pY = pY + (Cy - pY) / 2
      
plt.figure(figsize=(16,10), dpi= 80) 
plt.plot(x,y, 'bo')
plt.show()     
```
